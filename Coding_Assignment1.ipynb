```python
import pandas as pd
import pandas as pd
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
```

```python
tr = pd.read_csv("trial.csv")
aud_risk = pd.read_csv("audit_risk.csv")
```

```python
trialB = tr.drop(['Risk'], axis = 1)
```

```python
trialB.rename(columns = {'SCORE_A':'Score_A'}, inplace = True)
trialB.rename(columns = {'SCORE_B':'Score_B'}, inplace = True)
trialB['Score_A'] = trialB['Score_A']/10
trialB['Score_B'] = trialB['Score_B']/10
```

```python
intersected_columns = np.intersect1d(aud_risk.columns, trialB.columns)
intersected_columns
```

```python
combined_df = pd.merge(aud_risk, trialB, how='outer', on = ['History', 'LOCATION_ID', 'Money_Value', 'PARA_A', 'PARA_B',
       'Score', 'Score_A', 'Score_B', 'Sector_score', 'TOTAL', 'numbers'])
combined_df.columns
```

```python
combined_df.info()
```

```python
combined_df['Money_Value'] = combined_df['Money_Value'].fillna(combined_df['Money_Value'].median())
combined_df.describe()
```

```python
combined_df = combined_df.drop_duplicates(keep = 'first')
print("Total number of rows in the dataset: ",len(combined_df))
```

```python
combined_df["LOCATION_ID"].unique()
```

```python
combined_df = combined_df[(combined_df.LOCATION_ID != 'LOHARU')]
combined_df = combined_df[(combined_df.LOCATION_ID != 'NUH')]
combined_df = combined_df[(combined_df.LOCATION_ID != 'SAFIDON')]
print("Total number of rows in the dataset: ",len(combined_df))
```

```python
combined_df = combined_df.astype(float)
combined_df.info()
```

```python
import numpy as np


corr_matrix = combined_df.corr().abs()


upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))


columns = [column for column in upper.columns if any(upper[column] > 0.9)]

columns.append('Risk')
df = combined_df[columns]
```

```python
df.columns
```

```python
x = df.drop(['Risk'], axis = 1)
y = df['Risk']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.4, random_state = 0)
```

```python
dt = DecisionTreeClassifier(random_state = 0)
grid_param = {'max_depth': [3, 5, 7, 9, 11, 15]}
dt_grid = GridSearchCV(dt, grid_param, cv = 5, n_jobs = -1)
dt_grid.fit(x_train, y_train)
```

```python
print("Best Parameters for Decision Tree Classifier: ", dt_grid.best_params_)
```

```python
dt = DecisionTreeClassifier(max_depth = 5, random_state=0)
bag = BaggingClassifier(dt, random_state = 0)
grid_param = {'n_estimators': [100, 500, 1000],'max_samples': [0.1, 0.5, 1.0]}
bag_dt_grid = GridSearchCV(bag, grid_param, cv = 5, n_jobs = -1, return_train_score= True)
bag_dt_grid.fit(x_train, y_train)
```

```python
print("Best Parameters for Bagging Classifier: ", bag_dt_grid.best_params_)
```

```python
bag = BaggingClassifier(dt, n_estimators = 500, max_samples = 0.5, n_jobs = -1, bootstrap = False, random_state = 0)
bag.fit(x_train, y_train)
```

```python
print(" Training Score: ", bag.score(x_train, y_train))
print(" Testing Score: ", bag.score(x_test, y_test))
```
